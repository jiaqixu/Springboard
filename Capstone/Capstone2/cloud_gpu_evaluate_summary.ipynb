#### This is a summary of evalutation of train and trade for 13 datasets (results from cloud gpus)

```python
### Load modules
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Train Datasets

```python
# case 1

case1 = "Y"

def train_result(case1):
    if case1 =="Y":
        case = "case1"
    else:
        case = "case2"

    model_dict={}

    model_dict['loss']={}
    model_dict['acc']={}
    model_dict['roc']={}

    model_dict['loss']['mlp']={}
    model_dict['loss']['cnn']={}
    model_dict['loss']['rnn']={}
    model_dict['loss']['rnn_cnn']={}

    model_dict['acc']['mlp']={}
    model_dict['acc']['cnn']={}
    model_dict['acc']['rnn']={}
    model_dict['acc']['rnn_cnn']={}

    model_dict['roc']['mlp']={}
    model_dict['roc']['cnn']={}
    model_dict['roc']['rnn']={}
    model_dict['roc']['rnn_cnn']={}


    for ind in range(13):
        if ind <10:
            set_id = "0"+str(ind)
        else:
            set_id = str(ind)

        filename1 = '../models/'+case+'/trainHistory/mlp_'+case+'_train_set'+set_id+'.pkl'
        filename2 = '../models/'+case+'/trainHistory/cnn_'+case+'_train_set'+set_id+'.pkl'
        filename3 = '../models/'+case+'/trainHistory/rnn_'+case+'_train_set'+set_id+'.pkl'
        filename4 = '../models/'+case+'/trainHistory/rnn_cnn_'+case+'_train_set'+set_id+'.pkl'

        mlp_cur  = pickle.load(open(filename1,"rb"))
        cnn_cur = pickle.load(open(filename2,"rb"))
        rnn_cur = pickle.load(open(filename3,"rb"))
        rnn_cnn_cur = pickle.load(open(filename4,"rb"))

        model_dict['loss']['mlp'][ind] = round(mlp_cur['loss'][-1],4)
        model_dict['loss']['cnn'][ind] = round(cnn_cur['loss'][-1],4)
        model_dict['loss']['rnn'][ind] = round(rnn_cur['loss'][-1],4)
        model_dict['loss']['rnn_cnn'][ind] = round(rnn_cnn_cur['loss'][-1],4)

        model_dict['acc']['mlp'][ind] = round(mlp_cur['accuracy'][-1],4)
        model_dict['acc']['cnn'][ind] = round(cnn_cur['accuracy'][-1],4)
        model_dict['acc']['rnn'][ind] = round(rnn_cur['accuracy'][-1],4)
        model_dict['acc']['rnn_cnn'][ind] = round(rnn_cnn_cur['accuracy'][-1],4)

        model_dict['roc']['mlp'][ind] = round(mlp_cur['auc_roc'][-1],4)
        model_dict['roc']['cnn'][ind] = round(cnn_cur['auc_roc'][-1],4)
        model_dict['roc']['rnn'][ind] = round(rnn_cur['auc_roc'][-1],4)
        model_dict['roc']['rnn_cnn'][ind] = round(rnn_cnn_cur['auc_roc'][-1],4)



    train_loss = pd.DataFrame(model_dict['loss'])
    train_loss.index.set_names('ind',inplace=True)

    train_acc = pd.DataFrame(model_dict['acc'])
    train_acc.index.set_names('ind',inplace=True)

    train_roc = pd.DataFrame(model_dict['roc'])
    train_roc.index.set_names('ind',inplace=True)
    
    return train_loss, train_acc, train_roc
    #print(train_loss)
    #print(train_acc)
    #print(train_roc)
```

```python
index_year = ['2005/01-2008/01','2006/01-2009/01','2007/01-2010/01','2008/01-2011/01',
              '2009/01-2012/01','2010/01-2013/01','2011/01-2014/01','2012/01-2015/01',
              '2013/01-2016/01','2014/01-2017/01','2015/01-2018/01','2016/01-2019/01',
              '2017/01-2020/01']

train_loss, train_acc, train_roc = train_result(case1="Y")

train_acc.plot.bar(figsize = (30,5),rot=0, ylim = [0.5,1.1], fontsize=10)
#plt.bar(train_acc)
plt.xlabel("Train_set",fontsize=12)
plt.xticks(train_acc.index,index_year, fontsize =12)
plt.ylabel("Train_accuracy",fontsize=12)
plt.legend(prop={'size': 12})
plt.show()

train_loss.plot.bar(figsize = (30,5),rot=0, ylim = [0,0.2], fontsize=10)
#plt.bar(train_acc)
plt.xlabel("Train_set",fontsize=12)
plt.xticks(train_loss.index,index_year,fontsize =12)
plt.ylabel("Train_loss",fontsize=12)
plt.legend(prop={'size': 12})
plt.show()

train_roc.plot.bar(figsize = (30,5),rot=0, ylim = [0.7,1.1], fontsize=10)
#plt.bar(train_acc)
plt.xlabel("Train_set",fontsize=12)
plt.xticks(train_roc.index,index_year,fontsize =12)
plt.ylabel("Train_roc",fontsize=12)
plt.legend(prop={'size': 12})
plt.show()

#test_acc.plot.bar(figsize = (30,7),rot=0,ylim=[0.5,0.57],fontsize=10)
#plt.bar(train_acc)
#plt.xlabel("test_set",fontsize=20)
#plt.ylabel("test_accuracy",fontsize=20)
#plt.legend(prop={'size': 15})
#plt.show()
```

## Test Datasets

```python
### Read csv file
def test_result(case1):
    if case1 =="Y":
        case = "case1"
    else:
        case = "case2"

    filename1 = '../models/summary/'+case+'_loss.csv'
    filename2 = '../models/summary/'+case+'_acc.csv'
    filename3 = '../models/summary/'+case+'_roc.csv'


    test_loss = pd.read_csv(filename1,index_col='ind')
    test_acc = pd.read_csv(filename2,index_col='ind')
    test_roc = pd.read_csv(filename3,index_col='ind')
    
    return test_loss, test_acc, test_roc
```

```python
#case1

index_year = ['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020(b11)']
test_loss, test_acc, test_roc = test_result(case1="Y")

test_acc_p = test_acc[['mlp','cnn','rnn','rnn_cnn']]
test_acc_p.plot.bar(figsize = (30,5),rot=0, ylim = [0.3,0.8], fontsize=10)
plt.plot(test_acc.index,test_acc['stay'],'ro')
#plt.legend('stay',loc='upper left')
#plt.bar(train_acc)
plt.xlabel("Test_set",fontsize=12)
plt.xticks(test_acc_p.index,index_year, fontsize =12)
plt.ylabel("Test_accuracy",fontsize=12)
plt.legend(prop={'size': 12})

plt.show()

test_loss.plot.bar(figsize = (30,5),rot=0, ylim = [0,0.5], fontsize=10)
#plt.bar(train_acc)
plt.xlabel("Test_set",fontsize=12)
plt.xticks(test_loss.index,index_year,fontsize =12)
plt.ylabel("Test_loss",fontsize=12)
plt.legend(prop={'size': 12})
plt.show()

test_roc.plot.bar(figsize = (30,5),rot=0, ylim = [0.3,1], fontsize=10)
#plt.bar(train_acc)
plt.xlabel("Test_set",fontsize=12)
plt.xticks(test_roc.index,index_year,fontsize =12)
plt.ylabel("Test_roc",fontsize=12)
plt.legend(prop={'size': 12})
plt.show()
```
